```python
import cv2 
from google.colab.patches import cv2_imshow 
```

```python
imge = cv2.imread('a.jpeg', 0)
```

```python
cv2_imshow(imge)
```

```python
# 1 0 255 3
# 1 51 25 1
# 0 0 0 0 
# 61 0 7 1
```

```python
def turev_yatay(im, yatay, dikey):
  return im[dikey, yatay+1] - im[dikey, yatay] 

def turev_dikey(im, yatay, dikey):
  return im[dikey + 1, yatay] - im[dikey, yatay]
```

```python
print(turev_yatay(imge, 100, 100))
```

```python
# [-1, 1]
```

```python
kernel = np.array([[-1], [1]]).T

turev_yatay = cv2.filter2D(imge, -1, kernel)

kernel = np.array([[-1], [1]])

turev_dikey = cv2.filter2D(imge, -1, kernel)

cv2_imshow(turev_yatay)

cv2_imshow(turev_dikey)
```

```python
import math 

g = np.sqrt(turev_yatay**2 + turev_yatay**2)
```

```python
g = (255/np.max(g))*g

cv2_imshow(g)
```
